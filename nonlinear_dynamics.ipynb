```julia
f(x) = sqrt(x);
g(x) = -sqrt(x);

fig4 = plot([-3,3],[0,0], arrow=true,c=:black)
plot!([0,0], [-3,3],arrow=true,c=:black)
plot!(x->f(x),c=1)
plot!(x->g(x),c=2, ls=:dash)
annotate!(-0.25,-0.25,L"O")
annotate!(3-0.25,-0.25,L"\lambda")
annotate!(-0.25,3-0.25,L"\vec{x}_{0}")

for i in 1:5
    plot!([0.5i,0.5i],[2.9, f(0.5i) + 0.1],arrow=true,c=1)
end
for i in 1:5
    plot!([0.5i,0.5i],[g(0.5i) + 0.1, f(0.5i) - 0.1],arrow=true,c=2)
end
for i in 1:5
    plot!([0.5i,0.5i],[ g(0.5i) - 0.1,-2.9],arrow=true,c=2)
end
for i in 1:5
    plot!([-0.5i,-0.5i],[ 2.9,-2.9],arrow=true,c=3)
end

plot(fig4)

savefig(fig4, "分岔图-鞍节点分支.svg")
```

```julia

x,y = 0,1



startpoint = [1.5,0.1]
x = [startpoint[1]]
y = [startpoint[2]]
```

```julia
function dynamics(lambda1,lambda2,startpoint,num)
    x = [startpoint[1]]
    y = [startpoint[2]]
    for i in 1:num
        dx = lambda1 * 0.01 * x[i];
        dy = lambda2 * 0.01 * y[i];
        push!(x, x[i] + dx)
        push!(y, y[i] + dy)
    end
    return x,y
end
```

```julia
plot!(dynamics(-1,1,[3,3],100),arrow=true)
```

```julia
fig6 = plot([-3,3],[0,0], arrow=true,c=:black)
plot!([0,0], [-3,3],arrow=true,c=:black)
plot!(dynamics(-1,1,[2.9,0.1],300),arrow=true, c=2, ls=:dash)
plot!(dynamics(-1,1,[2.9,0.3],220),arrow=true,c=2, ls= :dash)
plot!(dynamics(-1,1,[-2.9,0.1],300),arrow=true,c=2,ls=:dash)
plot!(dynamics(-1,1,[-2.9,0.3],220),arrow=true,c=2, ls= :dash)
plot!(dynamics(-1,1,[-2.9,-0.1],300),arrow=true,c=2,ls=:dash)
plot!(dynamics(-1,1,[-2.9,-0.3],220),arrow=true,c=2, ls= :dash)
plot!(dynamics(-1,1,[2.9,-0.1],300),arrow=true,c=2,ls=:dash)
plot!(dynamics(-1,1,[2.9,-0.3],220),arrow=true,c=2, ls= :dash)
plot!(dynamics(-1,1,[2.9,0],300),arrow=true,c=1 )
plot!(dynamics(-1,1,[-2.9,0],300),arrow=true,c=1 )
plot!(dynamics(-1,1,[0,0.1],300),arrow=true,c=3 )
plot!(dynamics(-1,1,[0,-0.1],300),arrow=true,c=3 )
annotate!(2.9,-0.2, L"x'")
annotate!(-0.2,2.9, L"y'")
```

```julia
savefig(fig6, "相图-鞍点.svg")
```

```julia
fig7 = plot([-3,3],[0,0], arrow=true,c=:black)
plot!([0,0], [-3,3],arrow=true,c=:black)
teta = 0:0.01:3;
xf = sqrt.(teta) .* cos.(3teta);
yf = sqrt.(teta) .* sin.(3teta);
plot!(xf,yf, aspect=:equal, show=true,arrow=true,c=1)
savefig(fig7,"相图-不稳定焦点.svg")
```

```julia
fig7 = plot([-3,3],[0,0], arrow=true,c=:black)
plot!([0,0], [-3,3],arrow=true,c=:black)
plot!(dynamics(1,1,[0.1,0.1],320),arrow=true, c=2, ls=:dash)
plot!(dynamics(1,1,[-0.1,0.1],320),arrow=true, c=2, ls=:dash)
plot!(dynamics(1,1,[0.1,-0.1],320),arrow=true, c=2, ls=:dash)
plot!(dynamics(1,1,[-0.1,-0.1],320),arrow=true, c=2, ls=:dash)
plot!(dynamics(1,1,[0.1,-0],320),arrow=true, c=1)
plot!(dynamics(1,1,[-0.1,-0],320),arrow=true, c=1)
plot!(dynamics(1,1,[0,0.1],320),arrow=true, c=1)
plot!(dynamics(1,1,[0,-0.1],320),arrow=true, c=1)
annotate!(2.9,-0.2, L"x'")
annotate!(-0.2,2.9, L"y'")

savefig(fig7,"相图-不稳定结点.svg")
```
